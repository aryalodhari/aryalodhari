```python
# to find the maximum and minimum value of a given flattened array
# Name : Arya Lodhari
# enroll : 216270307064

import numpy as np
arr = np.array([[1, 5, 3],
 [9, 1, 6],
 [7, 8, 3]])
farr = arr.flatten()
maxv = np.max(farr)
minv = np.min(farr)
print("Maximum value:", maxv)
print("Minimum value:", minv)
```

```python
# to compute the mean, standard deviation, and variance of a given
#array along the second axis
# Name : Arya Lodhari
# enroll : 216270307064

arr = np.array([[8, 2, 6],
 [1, 5, 9],
 [4, 8, 2]])
mean_v = np.mean(arr, axis=1)
std_deviation_v = np.std(arr, axis=1)
variance_v = np.var(arr, axis=1)
print("Mean along the second axis:", mean_v)
print("Standard Deviation along the second axis:", std_deviation_v)
print("Variance along the second axis:", variance_v)
```
