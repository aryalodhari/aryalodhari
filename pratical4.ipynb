```python
# to calculate mean across dimension, in a 2D numpy array
# Name : Arya Lodhari
# enroll : 216270307064

import numpy as np
arr = np.arange(1,10).reshape(3,3)
print("Array : \n",arr)
c = np.mean(arr)
print("\nColoum vise mean : ",c)
r = np.mean(arr,axis=0)
print("Row vise mean : ",r)


```

```python
# to calculate the difference between neighboring elements, elementwise of a given array
# Name : Arya Lodhari
# enroll : 216270307064

import numpy as np
arr = np.arange(1,10).reshape(3,3)
print("Array : \n",arr)
print()
diff=np.diff(arr,axis=0)
print("Diff of neighbouring element\n",diff)
diff2=np.diff(arr,axis=1)
print("\nDiff of elements\n",diff2)
```
